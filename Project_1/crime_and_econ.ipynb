Since we have dataframes of all the counties for years 2011-2019, it's time to start making dataframes for each county

We will join other county data into these dataframes later

```python

county_dFrames = [sc_offenses_2011, sc_offenses_2012, sc_offenses_2013, sc_offenses_2014, sc_offenses_2015,
 sc_offenses_2016, sc_offenses_2017, sc_offenses_2018, sc_offenses_2019]

county_count = {}

for frame in county_dFrames:
    for county in frame["County"]:
        if county in county_count:
            county_count[county] += 1
        else:
            county_count[county] = 1

#print(county_dFrames[0].to_dict())

#combine all the dataframes into one
#remove counties that have less than a 9 value count in county_count
sc_offenses = pd.concat(county_dFrames)
sc_offenses = sc_offenses.dropna()
sc_offenses_full = sc_offenses[sc_offenses["County"].map(county_count) > 8]


#number of counties with > 8 in county_count
print(len(sc_offenses_full["County"].unique()))


```

We're going to focus on the counties that are consistently in all of the dataframes. If they have a value of 9 in the dictionary, they will be usd for study. All else, they will be dropped for this particular study



```python
personal_income = pd.read_csv("Data\Per capita personal income by county US Dept of Commerce 2011-2021.csv")
#only include the counties that have a value count of 9 or more
personal_income = personal_income[personal_income["County"].map(county_count) > 8]
personal_income = personal_income.dropna()


in_poverty = pd.read_csv("Data\SAIPE Population w percent in poverty 2011-2021_1.csv", encoding = 'latin1')

#take away all columns that are
#list all population columns
pop_columns = [col for col in in_poverty.columns if 'Percentage' not in col]
#make a new dataframe with only the population columns
in_poverty_pop = in_poverty[pop_columns]
in_poverty_pop = in_poverty_pop[in_poverty_pop["County"].map(county_count) > 8]
in_poverty_pop = in_poverty_pop.dropna()

per_columns = [col for col in in_poverty.columns if 'Population' not in col]
in_poverty_per = in_poverty[per_columns]
in_poverty_per = in_poverty_per[in_poverty_per["County"].map(county_count) > 8]
in_poverty_per = in_poverty_per.dropna()

in_poverty_pop


housing_est = pd.read_csv("Data\Estimated Number of Housing Units 2010-2019 (1).csv",encoding = 'latin1' )
housing_est = housing_est[housing_est["County"].map(county_count) > 8]
housing_est = housing_est.dropna()


med_house_income = pd.read_csv("Data\Median Household Income by County 2011-2021_3.csv", encoding = 'latin1')
med_house_income = med_house_income[med_house_income["County"].map(county_count) > 8]
med_house_income = med_house_income.dropna()



med_house_income.shape
```
