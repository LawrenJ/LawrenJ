```python
#load all the county lists
sc_offenses_2011 = pd.read_csv("Data/SC_County_Offenses_2011 - 11tbl10sc.csv")
sc_offenses_2011 = sc_offenses_2011.dropna()
sc_offenses_2011["Year"] = 2011 #added 'year' column for later data exploration

sc_offenses_2011.dtypes
#have to cast some of the column data into ints from strings

#cast each string(that contains a comma) into a float and then into an int
sc_offenses_2011["Arson"] = [int(float(str(row).replace(',',''))) for row in sc_offenses_2011["Arson"]]
sc_offenses_2011["Aggravated assault"] = [int(float(str(row).replace(',',''))) for row in sc_offenses_2011["Aggravated assault"]]
sc_offenses_2011["Larceny-theft"] = [int(float(str(row).replace(',',''))) for row in sc_offenses_2011["Larceny-theft"]]
sc_offenses_2011["Motor vehicle theft"] = [int(float(str(row).replace(',',''))) for row in sc_offenses_2011["Motor vehicle theft"]]
sc_offenses_2011["Property crime"] = [int(float(str(row).replace(',',''))) for row in sc_offenses_2011["Property crime"]]
sc_offenses_2011["Violent crime"] = [int(float(str(row).replace(',',''))) for row in sc_offenses_2011["Violent crime"]]
sc_offenses_2011["Burglary"] = [int(float(row.replace(',', ''))) for row in sc_offenses_2011["Burglary"]]

sc_offenses_2011.dtypes#checking if it worked
sc_offenses_2011.describe

```
